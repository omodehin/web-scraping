```python
# Importing the python Libaries

import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from random import randint
from time import sleep
```

```python
#Creating an empty list
ranke=[]
titlee=[]
certificatee=[]
duratione =[]
Genree=[]
Ratinge=[]
No_of_votee=[]
metae=[]
```

```python
 #My user Agent
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'}
```

```python
#using for loop to loop through all the pages of the website...the website has 9932 unique items with 50 items per page

for i in range(1,10000,50):
    link =' https://www.imdb.com/search/title/?title_type=feature&year=2020-01-01,2020-12-31&start='+str(i)+'&ref_=adv_nxt'
    #link='https://www.imdb.com/search/title/?year=2020&title_type=feature&'
    source =requests.get(link,headers=headers).text
    soup=BeautifulSoup(source,'html.parser') #creating the soup object....
    pages=soup.find_all('div', class_='lister-item mode-advanced')
    sleep(randint(2,8))
    for page in pages:
        try:
            ranke.append(page.find('h3',class_="lister-item-header").span.text)
        except Exception as e:
            ranke.append('NA')
        
        try:
            titlee.append(page.find('h3',class_="lister-item-header").a.text)
        except Exception as e:
            titlee.append('NA')
        try:
            certificatee.append(page.find('span',class_="certificate").string)
        except Exception as e:
            certificatee.append('NA')
        
        try:
            duratione.append(page.find('span',class_="runtime").text)
        except Exception as e:
            duratione.append('NA') 

        try:
            Genree.append(page.find('span',class_="genre").text.strip())
        except Exception as e:
            Genree.append('NA')

        try:
            Ratinge.append(page.find('div',class_="ratings-bar").strong.text)
        except Exception as e:
            Ratinge.append('NA')

        try:
            Votes=page.find_all('span', attrs = {'name':'nv'})
            vote =Votes[0].text
            No_of_votee.append(vote)
        except Exception as e:
            No_of_votee.append('NA')
            
        try:
            metae.append(page.find('span', class_='metascore').string)
        except Exception as e:
            metae.append('NA')
     
```

```python
#getting the series into python DataFrame
moviie =pd.DataFrame({'Rank':ranke,
                      'Title':titlee,
                      'Certificate':certificatee,
                      'Duration':duratione,
                      'Genre':Genree,
                      'Rating':Ratinge,
                      'Vote':No_of_votee,
                      'Metascore':metae})
```
