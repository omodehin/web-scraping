```python
#importing the libraries

import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from random import randint
from time import sleep
```

```python
#sending a request to the ncdc website
website = requests.get('https://covid19.ncdc.gov.ng/')
soup=BeautifulSoup(website.text,'lxml')
```

```python
#creating empty data for the rows
result=[]
```

```python
#creating empty data for headers
top=[]
```

```python
headers = soup.find('thead').find_all('tr')
for head in headers:
    top.append([hd.text for hd in head.find_all('th')])
```

```python
#finding the table body from the website
table = soup.find('tbody')
```

```python
#looping through the rows of the entire table

rows = table.find_all('tr')
for row in rows:
    result.append([data.text.strip() for data in row.find_all('td')])
```

```python
#output of the rows below
result
```

```python
df = pd.DataFrame(result, columns=top)
```
