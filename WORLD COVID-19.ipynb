```python
#IMPORTING THE NECESSARY PYTHON LIBRARIES

import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from random import randint
from time import sleep
```

```python
# WEBSITE LINK TO SCRAPE THE DATA FROM
link= 'https://www.worldometers.info/coronavirus/'
```

```python
website = requests.get(link)
```

```python
# CREATING THE SOUP OBJECT

soup = BeautifulSoup(website.content,'lxml')
```

```python
# CREATING EMPTY DATAFRAME FOR THE HEADERS AND THE ROWS

headers = []
rows = []
```

```python
all_columns =[]
```

```python
header = soup.find('table', id="main_table_countries_today")
head =header.find('thead').find('tr')
headers.append([data.text.replace('\n','') for data in head.find_all('th')])
 
```

```python
row_data = header.find_all('tr')
for row in row_data:
    all_columns.append([data.text.strip() for data in row.find_all('td')])
```

```python
new_data = all_columns[8:-5]
```

```python
#CONVERTING THE THE DATA INTO A DATAFRAME

df = pd.DataFrame(new_data,columns=headers)
```

```python
#CONVERTING THE DATAFRAME INTO AN EXCEL FILE

df.to_excel('WORLDCOVID_19.xlsx')
```
